```python
%%writefile guardar_datos.py
import os

# Ruta absoluta de salida en tu mÃ¡quina
BASE_OUTPUT = r"C:\Users\felip\Documents\Portafolio_Jupyter\data\processed"

def guardar_datos(datasets_dict):
    """
    Guarda DataFrames en BASE_OUTPUT sobrescribiendo siempre.
    datasets_dict: dict de nombre -> DataFrame
    Retorna dict de nombre -> ruta guardada
    """
    os.makedirs(BASE_OUTPUT, exist_ok=True)
    rutas_finales = {}
    for nombre, df in datasets_dict.items():
        ruta = os.path.join(BASE_OUTPUT, f"{nombre}_clean.csv")
        df.to_csv(ruta, index=False)
        rutas_finales[nombre] = ruta
    return rutas_finales
```
